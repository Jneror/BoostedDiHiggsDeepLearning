```python
#ada library
ada_parent_dir = "../"
import sys
sys.path.append(ada_parent_dir)
from ada import read_dataset, split_dataset_bytag, BinClassifModelV1 as BC1, plot_confidence_matrix
import pandas as pd
```

```python
source_path = "../../processed_data/xtohh"
signal = "Xtohh2000"
bg = "Xtohh_background"
seed = 420
```

```python
df = read_dataset(source_path, signal, bg, "SR", None)
sets = split_dataset_bytag(df, 0.6, 0.2, 0.2, seed)
```

```python
bc1 = BC1(sets["x"]["test"].shape[1])
bc1.load("../../saved_models", "paramtagv2_jun_2020")
```

# Tag 0

```python
x_test_tag0 = sets[0]["x_test"]
y_test_tag0 = sets[0]["y_test"]
w_test_tag0 = sets[0]["w_test"]
```

```python
eval = bc1.complete_evaluation(
    x_test_tag0, y_test_tag0, w_test_tag0, 0.4, save = True, name = "paramtag0_jun_2020"
)
```

## Tag 0: Non weighted evaluation

```python
print("Accuracy:", eval["accuracy"])
```

```python
pd.DataFrame(eval["class_report"])
```

```python
plot_confidence_matrix(eval["cm"], 6, 6, 'd')
```

## Tag 0: Weighted evaluation

```python
print("Accuracy:", eval["weighted_accuracy"])
```

```python
pd.DataFrame(eval["weighted_class_report"])
```

```python
plot_confidence_matrix(eval["weighted_cm"], 6, 6, '.1f')
```
