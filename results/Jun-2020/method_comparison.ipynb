```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
def plot_evaluation(tag, index, title, colors):
    df = tag.loc[index].drop(index = "support", level = 1)
    f, a = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

    df.xs('signal').plot(kind='bar',ax=a[0], color=colors)
    df.xs('bg').plot(kind='bar',ax=a[1], color=colors)

    a[0].get_legend().remove(); a[1].legend(bbox_to_anchor=(1.0, 1.2))
    a[0].title.set_text('Signal'); a[1].title.set_text('BG')

    f.suptitle(title, fontsize=20)
    f.autofmt_xdate(rotation=0, ha='center')

    a[0].grid(); a[1].grid()
    plt.show()
```

```python
#import evaluations
with open("eval_BCM1_tag0_jun_2020.json", 'r') as json_file:
    bc1_tag0 = json.loads(json_file.read())
    json_file.close()

with open("eval_BCM1_tag1_jun_2020.json", 'r') as json_file:
    bc1_tag1 = json.loads(json_file.read())
    json_file.close()

with open("eval_BCM1_tag2_jun_2020.json", 'r') as json_file:
    bc1_tag2 = json.loads(json_file.read())
    json_file.close()

with open("eval_BCM1_paramtag0_jun_2020.json", 'r') as json_file:
    bc1_paramtag0 = json.loads(json_file.read())
    json_file.close()

with open("eval_BCM1_paramtag1_jun_2020.json", 'r') as json_file:
    bc1_paramtag1 = json.loads(json_file.read())
    json_file.close()

with open("eval_BCM1_paramtag2_jun_2020.json", 'r') as json_file:
    bc1_paramtag2 = json.loads(json_file.read())
    json_file.close()

with open("eval_AM1_tag2_jun_2020.json", 'r') as json_file:
    a1_tag2 = json.loads(json_file.read())
    json_file.close()
```

# Tag 0

```python
tag0 = pd.concat({
    "non-weighted": pd.concat({
        "signal": pd.DataFrame({
            'BC1': bc1_tag0['class_report']['1'],
            'Parameterized-BC1 ': bc1_paramtag0['class_report']['1'],
        }),
        "bg": pd.DataFrame({
            'BC1': bc1_tag0['class_report']['0'],
            'Parameterized-BC1 ': bc1_paramtag0['class_report']['0'],
        })
    }),
    "weighted": pd.concat({
        "signal": pd.DataFrame({
            'BC1': bc1_tag0['weighted_class_report']['1'],
            'Parameterized-BC1 ': bc1_paramtag0['weighted_class_report']['1'],
        }),
        "bg": pd.DataFrame({
            'BC1': bc1_tag0['weighted_class_report']['0'],
            'Parameterized-BC1 ': bc1_paramtag0['weighted_class_report']['0'],
        })
    })
})
tag0
```

```python
plot_evaluation(tag0, "non-weighted", "Tag 0 non-weighted results", ["#364f6b", "#3fc1c9"])
```

```python
plot_evaluation(tag0, "weighted", "Tag 0 weighted results", ["#364f6b", "#3fc1c9"])
```
