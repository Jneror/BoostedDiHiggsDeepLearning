```python
import pandas as pd
import os
from ROOT import TFile
from root_numpy import tree2array
from IPython.display import display
```

### read root files

```python
def read_root_file(file):
    rfile = TFile(file) #root file
    intree = rfile.Get("Nominal") #get tree from root file
    array = tree2array(intree)
    df = pd.DataFrame(array)
    return df
```

```python
root_files = os.listdir("../raw_data/")
regions = ["SR", "QCDCR","TwoDimMassWindow"]
tags = [0,1,2]
d_reg_tag = {}
for reg in regions:
    for tag in tags:
        d_samples = {}
        print("-"*5,reg,"tag",tag)
        for rf in root_files:
            df = read_root_file("../raw_data/"+rf)
            m1 = df["m_region"] == reg
            m2 = df['m_FJNbtagJets'] == tag
            df = df[ m1 & m2]
            d = df["sample"].value_counts()
            for samp in d.keys():
                if samp not in d_samples:
                    d_samples[samp]=d[samp]
                else:
                    d_samples[samp]+=d[samp]
        print(d_samples)
        d_reg_tag[reg+"_"+str(tag)] =pd.DataFrame.from_dict(d_samples,orient='index', columns=['entries'])
           
```
