```python
import pandas as pd
from glob import glob
ada_parent_dir = "../"
import sys
sys.path.append(ada_parent_dir)
import ada
```

```python
#import datasets
data_path = "../processed_data/"
signal = "Xtohh"
```

```python
signal_paths = glob(data_path + signal + "[!(_background)]*.csv")
```

```python
#Loading background
bg_df = pd.read_csv(data_path + signal + "_background.csv")
```

```python
bg_df
```

```python
#loading signal csv's into a dictionary of dataframes
signal_dfs = {signal_path.split('/')[-1][:-4]: pd.read_csv(signal_path) for signal_path in signal_paths}
```

```python
#drop fakes
signal_dfs = {name: ada.drop_fakes(df) for name, df in signal_dfs.items()}
bg_df = ada.drop_fakes(bg_df)
```

```python
#getting background info
bg_weight = bg_df["EventWeight"].sum()
bg_count = bg_df.shape[0]
bg_weight, bg_count
```

```python
signals_info = {name: {'weight': df["EventWeight"].sum() , 'count': df.shape[0]} for name, df in signal_dfs.items()}
```

```python
singals_info_df = pd.DataFrame(signals_info).T
```

```python
singals_info_df
```

```python
singals_info_df["density"] = singals_info_df["count"] / (bg_count + singals_info_df["count"])
```

```python
singals_info_df
```

```python
singals_info_df["weighted_density"] = singals_info_df["weight"] / (bg_weight + singals_info_df["weight"])
```
