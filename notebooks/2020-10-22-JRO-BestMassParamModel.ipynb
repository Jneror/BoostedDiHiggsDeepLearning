```python
# Ada library
ada_parent_dir = "../"
import sys
sys.path.append(ada_parent_dir)
from ada2.data import read_dataset, split_dataset_bymass_2, all_hyperparams_comb
from ada2.model import FeedForwardNeuralNetworkArchitecture as FFNNA
from ada2.plot import plot_confidence_matrix
import pandas as pd
from IPython.core.display import display, HTML

# Keras
from keras.optimizers import adam, adadelta, adagrad, adamax
```

```python
# Constants
source_path = "../processed_data/xtohhOct2020"
masses = [1000, 1200, 1400, 1600, 1800, 2000, 2500, 3000]
signal = "Xtohh"
bg = "Xtohh_background"
seed = 420
```

```python
# Read xtohh dataset: SR region and tag 2
dfs = {mass: read_dataset(source_path, signal+str(mass), bg, "SR", 2) for mass in masses}
df = pd.concat(dfs, names = ["mass", None])
df = df.reset_index(level = 0).reset_index(drop = True)
df
```

```python
sets = split_dataset_bymass_2(df, 0.5, 0.3, 0.2, seed, masses)
```

```python
model = FFNNA(sets["x"]["train"].shape[1], 0.05, adamax, "relu")
model.fit(
    sets["x"]["train"], sets["y"]["train"], sets["w"]["train"],
    sets["x"]["val"], sets["y"]["val"], sets["w"]["val"],
    50, verbose = 1
)
model.plot_loss()
```

```python
def scores_per_mass(model, sets, masses, th):
    for mass in masses:
        display(HTML(f"<h2>Xtohh {mass} Tag 2</h2>"))
        x_test_mass = sets[mass]["x_test"]
        y_test_mass = sets[mass]["y_test"]
        w_test_mass = sets[mass]["w_test"]

        eval = model.complete_evaluation(
            x_test_mass, y_test_mass, w_test_mass, th, save = False
        )

        print("Accuracy:", eval["weighted_accuracy"])
        display(pd.DataFrame(eval["weighted_class_report"]))
        plot_confidence_matrix(eval["weighted_cm"], 6, 6, '.1f')
```

```python
scores_per_mass(model, sets, masses, 0.7)
```

```python
model2 = FFNNA(sets["x"]["train"].shape[1], 0.01, adamax, "relu")
model2.fit(
    sets["x"]["train"], sets["y"]["train"], sets["w"]["train"],
    sets["x"]["val"], sets["y"]["val"], sets["w"]["val"],
    100, verbose = 1
)
model2.plot_loss()
```

```python
scores_per_mass(model2, sets, masses, 0.7)
```

```python
def scores_per_mass_no_negative_w(model, sets, masses, th):
    for mass in masses:
        display(HTML(f"<h2>Xtohh {mass} Tag 2</h2>"))
        x_test_mass, y_test_mass, w_test_mass = (sets[mass]["x_test"], sets[mass]["y_test"], sets[mass]["w_test"])
        x_test_mass = x_test_mass[w_test_mass >= 0.]
        y_test_mass = y_test_mass[w_test_mass >= 0.]
        w_test_mass = w_test_mass[w_test_mass >= 0.]

        eval = model.complete_evaluation(
            x_test_mass, y_test_mass, w_test_mass, th, save = False
        )

        print("w-Accuracy:", eval["weighted_accuracy"])
        display(pd.DataFrame(eval["weighted_class_report"]))
        plot_confidence_matrix(eval["weighted_cm"], 6, 6, '.1f')
        plot_confidence_matrix(eval["cm"], 6, 6, 'd')
```

```python
scores_per_mass_no_negative_w(model, sets, masses, 0.7)
```

```python
scores_per_mass_no_negative_w(model2, sets, masses, 0.7)
```
